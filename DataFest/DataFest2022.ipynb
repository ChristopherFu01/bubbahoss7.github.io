```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("player-6607011.csv")
```

```python
cols=['skill_level_know','skill_level_priority',"skill_level_refusal","skill_level_me","skill_level_people","event_time_dbl","event_id"]
df=df[cols]
```

```python
df=df.fillna(method="ffill")
```

```python
df["skill_level_refusal"]
```

```python
df['skill_level_people'].unique()
```

```python
fig,ax=plt.subplots(figsize=(12,8))
sns.lineplot(x="event_time_dbl", y="skill_level_refusal",
             data=df)
sns.lineplot(x="event_time_dbl", y="skill_level_me",
             data=df)
sns.lineplot(x="event_time_dbl", y="skill_level_priority",
             data=df)
sns.lineplot(x="event_time_dbl", y="skill_level_people",
             data=df)
sns.lineplot(x="event_time_dbl", y="skill_level_know",
             data=df)
```
